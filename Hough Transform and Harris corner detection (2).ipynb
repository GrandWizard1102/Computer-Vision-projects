```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image=cv2.imread(r'C:\Users\student\Desktop\CV20\square.jpg')
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap='gray')
plt.axis('off')
```

```python
edges=cv2.Canny(gray,50,100,apertureSize=3)
plt.imshow(edges,cmap='gray')
plt.axis('off')
```

```python
lines=cv2.HoughLinesP(edges,1,np.pi/180,100,50,10)
```

```python
lines.shape
```

```python
if lines is not None:
    for line in lines:
        x1,y1,x2,y2=line[0]
        
        cv2.line(image,(x1,y1),(x2,y2),(0,0,255),2)
```

```python
plt.imshow(image,cmap='gray')
plt.axis('off')
```

```python
dst=cv2.cornerHarris(gray,2,3,0.04)
```

```python
dst=cv2.dilate(dst,None)
```

```python
image[dst>0.01*dst.max()]=(255,0,0)
```

```python
plt.imshow(image,cmap='gray')
plt.axis('off')
```

```python
segment=np.zeros_like(image)
```

```python
if lines is not None:
    for line in lines:
        x1,y1,x2,y2=line[0]
        
        cv2.line(segment,(x1,y1),(x2,y2),(0,0,255),2)
```

```python
segmented = cv2.bitwise_and(image,segment)
```

```python
plt.imshow(segmented)
plt.axis('off')
```
