```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image=cv2.imread(r'C:\Users\student\Desktop\CV20\sample.jpg')
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
plt.axis('off')
plt.title("RGB Model")
```

```python
image_red=np.zeros_like(image,dtype=float)
image_red[:,:,0]=image[:,:,0]
image_red/=255.0

plt.figure(figsize=(8,5))
plt.subplot(1,3,1)
plt.axis('off')
plt.imshow(image_red)
plt.title('Red channel')

image_green=np.zeros_like(image,dtype=float)
image_green[:,:,1]=image[:,:,1]
image_green/=255.0
plt.subplot(1,3,2)
plt.axis('off')
plt.imshow(image_green)
plt.title('Green channel')

image_blue=np.zeros_like(image,dtype=float)
image_blue[:,:,2]=image[:,:,2]
image_blue/=255.0
plt.subplot(1,3,3)
plt.imshow(image_blue)
plt.axis('off')
plt.title('Blue channel')

plt.tight_layout()
```

```python
image.shape
```

```python
image_cmy=np.zeros_like(image,dtype=float)
image_cmy[:,:,0]=1-(image[:,:,0]/255.0)
image_cmy[:,:,1]=1-(image[:,:,1]/255.0)
image_cmy[:,:,2]=1-(image[:,:,2]/255.0)

image_cmy=(image_cmy*255).astype(np.uint8)
```

```python
plt.imshow(image_cmy)
plt.axis('off')
plt.title("CMY Model")
```

```python
image_yiq=np.zeros_like(image,dtype=float)
image_yiq[:,:,0]=0.299*image[:,:,0]+0.587*image[:,:,1]+0.114*image[:,:,2]
image_yiq[:,:,1]=0.596*image[:,:,0]-0.275*image[:,:,1]-0.321*image[:,:,2]
image_yiq[:,:,2]=0.212*image[:,:,0]-0.523*image[:,:,1]+0.311*image[:,:,2]

image_yiq[:,:,0]=np.clip( image_yiq[:,:,0],0,255)
image_yiq[:,:,1]=np.clip( image_yiq[:,:,1]+128,0,255)
image_yiq[:,:,2]=np.clip( image_yiq[:,:,2]+128,0,255)

image_yiq = image_yiq.astype(np.uint8)
```

```python
plt.imshow(image_yiq)
plt.axis('off')
plt.title("YIQ Model")
```

```python
image_hsv=cv2.cvtColor(image,cv2.COLOR_RGB2HSV)


image_hue=image_hsv[:,:,0].astype(float)
image_hue/=179.0

image_saturation=image_hsv[:,:,1].astype(float)


image_value=image_hsv[:,:,2].astype(float)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.imshow(image_hsv)
plt.axis('off')
plt.title("HSV Model")

plt.subplot(2,2,2)
plt.imshow(image_hue)
plt.axis('off')
plt.title("Hue channel")

plt.subplot(2,2,3)
plt.imshow(image_saturation,cmap='gray')
plt.axis('off')
plt.title("Saturation channel")

plt.subplot(2,2,4)
plt.imshow(image_value,cmap='gray')
plt.axis('off')
plt.title("Value channel")
```
